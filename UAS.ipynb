```python
import pandas as pd
import numpy as np
from numpy import array
df = pd.read_csv('https://raw.githubusercontent.com/DwiAqilahP/kolaborasipro/main/PTRO.JK.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Open'] = df['Open'].fillna(value=df['Open'].median())
df['High'] = df['High'].fillna(value=df['High'].median())
df['Low'] = df['Low'].fillna(value=df['Low'].median())
df['Close'] = df['Close'].fillna(value=df['Close'].median())
df['Adj Close'] = df['Adj Close'].fillna(value=df['Adj Close'].median())
df['Volume'] = df['Volume'].fillna(value=df['Volume'].median())
```

```python
df.isnull().sum()
```

```python
data = df['Open']
data
```

```python
n = len(data)
sizeTrain = (round(n*0.8))
data_train = pd.DataFrame(data[:sizeTrain])
data_test = pd.DataFrame(data[sizeTrain:])
# dates_test = pd.DataFrame(dates[sizeTrain:])
data_train
```

```python
# Normalisasi data menggunakan MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(data_train)


# sc = MinMaxScaler(feature_range = (0, 1))
# training_set_scaled = sc.fit_transform(data_train)

# Mengaplikasikan MinMaxScaler pada data pengujian
test_scaled = scaler.transform(data_test)

# reshaped_data = data.reshape(-1, 1)
train = pd.DataFrame(train_scaled, columns = ['data'])
train = train['data']

test = pd.DataFrame(test_scaled, columns = ['data'])
test = test['data']
test


```

```python
def split_sequence(sequence, n_steps):
  X, y = list(), list()
  for i in range(len(sequence)):
    # find the end of this pattern
    end_ix = i + n_steps
    # check if we are beyond the sequence
    if end_ix > len(sequence)-1:
      break
    # gather input and output parts of the pattern
    seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
    X.append(seq_x)
    y.append(seq_y)

  return array(X), array(y)
```

```python
X, Y = split_sequence(train, 2)
x = pd.DataFrame(X, columns = ['xt-2','xt-1'])
y = pd.DataFrame(Y, columns = ['xt'])
dataset_train = pd.concat([x, y], axis=1)
dataset_train
# dataset_train.to_excel('data_train.xlsx', index=False)
```

```python
X_train = dataset_train.iloc[:, :2].values
Y_train = dataset_train.iloc[:, -1].values
X_train
```

```python
test_x, test_y = split_sequence(test, 2)
x = pd.DataFrame(test_x, columns = ['xt-2','xt-1'])
y = pd.DataFrame(test_y, columns = ['xt'])
dataset_test = pd.concat([x, y], axis=1)
dataset_test
# dataset_test.to_excel('data-test.xlsx', index=False)
```

```python
X_test = dataset_test.iloc[:, :2].values
Y_test = dataset_test.iloc[:, -1].values
print(len(X_test))
```

```python
# Melakukan fitting dan prediksi menggunakan model KNeighborsRegressor
from sklearn.neighbors import KNeighborsRegressor
knn_model = KNeighborsRegressor(n_neighbors=6)
knn_model.fit(X_train, Y_train)
knn_preds = knn_model.predict(X_test)

# Melakukan fitting dan prediksi menggunakan model DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, Y_train)
dt_preds = dt_model.predict(X_test)

# Melakukan fitting dan prediksi menggunakan model RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(X_train, Y_train)
rf_preds = rf_model.predict(X_test)

# Melakukan fitting dan prediksi menggunakan model LinearRegression
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)
lr_preds = lr_model.predict(X_test)

# Melakukan fitting dan prediksi menggunakan model SVR
from sklearn.svm import SVR
svr_model = SVR()
svr_model.fit(X_train, Y_train)
svr_preds = svr_model.predict(X_test)
```

```python
# Mengembalikan data hasil prediksi ke skala semula
knn_preds = scaler.inverse_transform(knn_preds.reshape(-1, 1)).flatten()
dt_preds = scaler.inverse_transform(dt_preds.reshape(-1, 1)).flatten()
rf_preds = scaler.inverse_transform(rf_preds.reshape(-1, 1)).flatten()
lr_preds = scaler.inverse_transform(lr_preds.reshape(-1, 1)).flatten()
svr_preds = scaler.inverse_transform(svr_preds.reshape(-1, 1)).flatten()
```

```python
# Menyimpan hasil prediksi dan nilai aktual ke dalam file Excel
prediksi_knn = pd.DataFrame({'Prediksi': knn_preds})
# prediksi.to_excel('prediksi_knn.xlsx', index=False)

prediksi_dt = pd.DataFrame({'Prediksi': dt_preds})
# prediksi.to_excel('prediksi_dt.xlsx', index=False)

prediksi_rf = pd.DataFrame({'Prediksi': rf_preds})
# prediksi.to_excel('prediksi_rf.xlsx', index=False)

prediksi_lr = pd.DataFrame({'Prediksi': lr_preds})
# prediksi.to_excel('prediksi_lr.xlsx', index=False)

prediksi_svr = pd.DataFrame({'Prediksi': svr_preds})
# prediksi.to_excel('prediksi_svr.xlsx', index=False)

aktual = pd.DataFrame({'Aktual': scaler.inverse_transform(test_y.reshape(-1, 1)).flatten()})
# aktual.to_excel('aktual.xlsx', index=False)
```

```python
# Menghitung Mean Absolute Percentage Error (MAPE) untuk setiap model
from sklearn.metrics import mean_absolute_percentage_error
knn_mape = mean_absolute_percentage_error(aktual, knn_preds) * 100
dt_mape = mean_absolute_percentage_error(aktual, dt_preds) * 100
rf_mape = mean_absolute_percentage_error(aktual, rf_preds) * 100
lr_mape = mean_absolute_percentage_error(aktual, lr_preds) * 100
svr_mape = mean_absolute_percentage_error(aktual, svr_preds) * 100
```
